# Extracting the data

```python
pitches = []
durations = []

for i in first_score.flat:
    
    if isinstance(i, chord.Chord):
        pitches.append('.'.join(n.nameWithOctave for n in i.pitches))
        durations.append(i.duration.quarterLength)

    if isinstance(i, note.Note):
        if i.isRest:
            pitches.append(str(i.name))
            durations.append(i.duration.quarterLength)
        else:
            pitches.append(str(i.nameWithOctave))
            durations.append(i.duration.quarterLength)

    
```

```python
print('\nDurations', 'Pitches')
for pi,du in zip(pitches,durations):
    print(du, '\t', pi)
```

```python
#Create a new track from scratch

from mido import MidiTrack, MidiFile, Message

bpm = 30
def play(note, length_of_the_track, track,base=0,delayed_by=0,played_velocity=1.5,channel=0):
    meta_time = 60 * 60 * 10 / bpm
   
    noted=1
    major_notes = [1, 2, 2, 2, 2, 1, 0, 1]
    base_note = 50
    track.append(
        Message('note_on', note=base_note + base * 12 *noted+ sum(major_notes[0:note])*noted, velocity=round(78 * played_velocity),
                time=round(delayed_by * meta_time), channel=channel))
    track.append(
        Message('note_off', note=base_note + base * 14 *noted + sum(major_notes[0:note])*noted, velocity=round(78 * played_velocity),
                time=round(meta_time * length_of_the_track), channel=channel))
 
 
def sentences(track):
    
    play(1, 1.5, track) 
    play(2, 0.5, track,-2) 
    play(7, 0.25, track, -1) 
    play(6, 0.25, track, -1)  
    play(5, 0.5, track, -1, channel=1) 
    play(3, 0.5, track,-3, channel=1) 
    play(3, 2, track, channel=1) 
    play(3, 0.87,track) 
    play(4, 0.95,track)
    play(3, 1.567,track) 
    play(1, 0.25,track)
    play(7, 0.50, track) 
    play(6, 0.5, track, -1, channel=1) 
    play(3, 0.5, track, channel=1)
    play(9, 2, track, channel=1) 
    play(1, 0.65, track, -1)
    play(5, 0.57, track)
    play(8, 2.25, track, -1)
    play(3, 0.35, track, -1)
    play(7, 0.25, track, -1)
    play(5, 0.5, track, -1, channel=1) 
    play(2, 0.5, track, channel=1)
    play(2, 2.34, track, channel=1) 
    play(5, 1.5, track) 
    play(1, 0.5, track) 
    play(4, 0.5, track) 
    play(8, 0.5, track, -1)
 
    play(1, 3, track) 
 
def group(track, num):
        play(5, 0.5, track)
        play(6, 0.5, track)
        play(5, 1.5, track)
        play(3, 0.5, track) 
 
        play(5, 0.5, track, channel=1) 
        play(6, 0.5, track, channel=1)
        play(5, 2, track, channel=1)
 
        play(6, 0.5, track) 
        play(5, 0.5, track)
        play(4, 0.5, track) 
        if num == 1:
                play(1, 0.25, track, channel=1) 
                play(1, 0.25, track, channel=1)
        if num == 2:
                 play(1, 0.5, track, channel=1)
                 play(6, 0.5, track, channel=1) 
        play(5, 0.5, track, channel=1)
 
        play(5, 3, track, channel=1) 
 
        play(3, 0.5, track) 
        play(4, 0.5, track) 
        play(3, 1.5, track) 
        play(2, 0.25, track)
        play(1, 0.25, track)
 
        play(6, 0.5, track, -1, channel=1) 
        play(2, 0.5, track, channel=1) 
        play(2, 2, track, channel=1) 
 
        play(4, 0.5, track) 
        play(5, 0.5, track) 
        play(4, 0.5, track)
        play(3, 0.5, track)
        play(1, 0.5, track, channel=1) 
        play(6, 0.5, track, -1, channel=1)
 
        play(1, 3, track, channel=1) 
 
 # Creating MidiFile track and adding a audio track to it.
mid = MidiFile()
track = MidiTrack()
mid.tracks.append(track)
 #Adding the message to the tracks
track.append(Message('note_off', note=64, velocity=127, time=32))
track.append(Message('program_change', program=11))  
track.append(Message('note_on', note=64, velocity=64, time=32))

 
sentences(track) # verse

group(track, num=2) # chorus 2
group(track, num=1) # chorus 1
mid.save('new_song.mid')
```
